```python
from abc import ABC, abstractmethod
import json
from statistics import mean

# ---------- Soyut SÄ±nÄ±f ----------
class Medya(ABC):
    def __init__(self, baslik, yonetmen, tur, yil):
        self.baslik = baslik
        self.yonetmen = yonetmen
        self.tur = tur
        self.yil = yil
        self._puanlar = []  # KullanÄ±cÄ±larÄ±n verdiÄŸi puanlar

    @abstractmethod
    def bilgileri_goster(self):
        """Alt sÄ±nÄ±flarda override edilir."""
        pass

    def puan_ver(self, puan):
        """KullanÄ±cÄ±nÄ±n filme puan vermesini saÄŸlar."""
        try:
            puan = float(puan)
            if 0 <= puan <= 10:
                self._puanlar.append(puan)
                print(f"â­ '{self.baslik}' iÃ§in puan eklendi: {puan}")
            else:
                print("âš  Puan 0 ile 10 arasÄ±nda olmalÄ±.")
        except ValueError:
            print("âš  GeÃ§erli bir sayÄ± giriniz.")

    @property
    def ortalama_puan(self):
        """Ortalama puanÄ± hesaplar."""
        return round(mean(self._puanlar), 2) if self._puanlar else None

    def __str__(self):
        puan = f"{self.ortalama_puan}/10" if self.ortalama_puan else "Yok"
        return f"{self.baslik} ({self.yil}) | TÃ¼r: {self.tur} | Puan: {puan}"


# ---------- Alt SÄ±nÄ±flar ----------
class Film(Medya):
    def __init__(self, baslik, yonetmen, tur, yil, sure):
        super().__init__(baslik, yonetmen, tur, yil)
        self.sure = sure

    def bilgileri_goster(self):
        puan = f"{self.ortalama_puan}/10" if self.ortalama_puan else "HenÃ¼z yok"
        return (f"ðŸŽ¬ Film: {self.baslik} ({self.yil}) | TÃ¼r: {self.tur} | "
                f"SÃ¼re: {self.sure} dk | YÃ¶netmen: {self.yonetmen} | Puan: {puan}")


class Dizi(Medya):
    def __init__(self, baslik, yonetmen, tur, yil, sezon, bolum):
        super().__init__(baslik, yonetmen, tur, yil)
        self.sezon = sezon
        self.bolum = bolum

    def bilgileri_goster(self):
        puan = f"{self.ortalama_puan}/10" if self.ortalama_puan else "HenÃ¼z yok"
        return (f"ðŸ“º Dizi: {self.baslik} ({self.yil}) | {self.sezon} Sezon, {self.bolum} BÃ¶lÃ¼m | "
                f"TÃ¼r: {self.tur} | YÃ¶netmen: {self.yonetmen} | Puan: {puan}")


# ---------- YardÄ±mcÄ± SÄ±nÄ±flar ----------
class Yonetmen:
    def __init__(self, ad, ulke):
        self.ad = ad
        self.ulke = ulke

    def __str__(self):
        return f"{self.ad} ({self.ulke})"


class Kullanici:
    def __init__(self, ad):
        self.ad = ad

    def film_puanla(self, medya, puan):
        medya.puan_ver(puan)


# ---------- Ana ArÅŸiv SÄ±nÄ±fÄ± ----------
class Arsiv:
    def __init__(self):
        self.medya_listesi = []

    def ekle(self, medya):
        self.medya_listesi.append(medya)

    def listele(self):
        print("\nðŸŽž Mevcut ArÅŸiv:")
        if not self.medya_listesi:
            print("HenÃ¼z hiÃ§bir medya eklenmemiÅŸ.")
            return
        for i, medya in enumerate(self.medya_listesi, start=1):
            print(f"{i}. {medya}")

    def ara(self, kelime):
        sonuc = [m for m in self.medya_listesi if kelime.lower() in m.baslik.lower()]
        return sonuc

    # --- JSON KayÄ±t / YÃ¼kleme (gerÃ§ek nesnelerle) ---
    def kaydet(self, dosya_adi="arsiv.json"):
        try:
            veri = []
            for m in self.medya_listesi:
                medya_dict = {
                    "tip": m.__class__.__name__,
                    "baslik": m.baslik,
                    "yonetmen": {"ad": m.yonetmen.ad, "ulke": m.yonetmen.ulke},
                    "tur": m.tur,
                    "yil": m.yil,
                    "_puanlar": m._puanlar
                }

                if isinstance(m, Film):
                    medya_dict["sure"] = m.sure
                elif isinstance(m, Dizi):
                    medya_dict["sezon"] = m.sezon
                    medya_dict["bolum"] = m.bolum

                veri.append(medya_dict)

            with open(dosya_adi, "w", encoding="utf-8") as f:
                json.dump(veri, f, ensure_ascii=False, indent=4)
            print("ðŸ’¾ ArÅŸiv baÅŸarÄ±yla kaydedildi.")
        except Exception as e:
            print("âŒ KayÄ±t hatasÄ±:", e)

    def yukle(self, dosya_adi="arsiv.json"):
        try:
            with open(dosya_adi, "r", encoding="utf-8") as f:
                veri = json.load(f)

            self.medya_listesi = []
            for m in veri:
                yonetmen = Yonetmen(m["yonetmen"]["ad"], m["yonetmen"]["ulke"])
                if m["tip"] == "Film":
                    medya = Film(m["baslik"], yonetmen, m["tur"], m["yil"], m["sure"])
                else:
                    medya = Dizi(m["baslik"], yonetmen, m["tur"], m["yil"], m["sezon"], m["bolum"])
                medya._puanlar = m["_puanlar"]
                self.medya_listesi.append(medya)

            print("ðŸ“‚ ArÅŸiv baÅŸarÄ±yla yÃ¼klendi.")
        except FileNotFoundError:
            print("âš  HenÃ¼z kayÄ±tlÄ± bir arÅŸiv bulunamadÄ±.")


# ---------- Konsol ArayÃ¼zÃ¼ ----------
def menu():
    arsiv = Arsiv()
    arsiv.yukle()  # Program aÃ§Ä±ldÄ±ÄŸÄ±nda varsa verileri yÃ¼kle
    kullanici = Kullanici("Zeynep")  # Ã–rnek kullanÄ±cÄ±

    while True:
        print("""
ðŸŽ¬ Film/Dizi ArÅŸiv Sistemi
1. Yeni film ekle
2. Yeni dizi ekle
3. ArÅŸivi listele
4. BaÅŸlÄ±ÄŸa gÃ¶re ara
5. Medyaya puan ver
6. En yÃ¼ksek puanlÄ± 5 medya
7. ArÅŸivi kaydet
8. Ã‡Ä±kÄ±ÅŸ
""")
        secim = input("SeÃ§iminiz: ")

        if secim == "1":
            baslik = input("Film adÄ±: ")
            yonetmen = Yonetmen(input("YÃ¶netmen adÄ±: "), input("Ãœlke: "))
            tur = input("TÃ¼r: ")
            yil = int(input("YÄ±l: "))
            while True:
                try:
                    sure = int(input("SÃ¼re (yalnÄ±zca dakika cinsinden sayÄ±): "))
                    break
                except ValueError:
                    print("âš  GeÃ§ersiz giriÅŸ. LÃ¼tfen sÃ¼reyi yalnÄ±zca dakika cinsinden sayÄ± olarak girin.")

            film = Film(baslik, yonetmen, tur, yil, sure)
            arsiv.ekle(film)

        elif secim == "2":
            baslik = input("Dizi adÄ±: ")
            yonetmen = Yonetmen(input("YÃ¶netmen adÄ±: "), input("Ãœlke: "))
            tur = input("TÃ¼r: ")
            yil = int(input("YÄ±l: "))
            sezon = int(input("Sezon sayÄ±sÄ±: "))
            bolum = int(input("BÃ¶lÃ¼m sayÄ±sÄ±: "))
            dizi = Dizi(baslik, yonetmen, tur, yil, sezon, bolum)
            arsiv.ekle(dizi)

        elif secim == "3":
            arsiv.listele()

        elif secim == "4":
            kelime = input("Aranacak kelime: ")
            sonuc = arsiv.ara(kelime)
            if sonuc:
                for m in sonuc:
                    print(m.bilgileri_goster())
            else:
                print("ðŸ” SonuÃ§ bulunamadÄ±.")

        elif secim == "5":
            arsiv.listele()
            try:
                no = int(input("Puan vermek istediÄŸiniz medya numarasÄ±: ")) - 1
                puan = input("Puan (0-10): ")
                kullanici.film_puanla(arsiv.medya_listesi[no], puan)
            except (ValueError, IndexError):
                print("âš  GeÃ§ersiz seÃ§im.")

        elif secim == "6":
            sirali = sorted(
                [m for m in arsiv.medya_listesi if m.ortalama_puan],
                key=lambda x: x.ortalama_puan,
                reverse=True
            )
            print("\nðŸ† En yÃ¼ksek puanlÄ± 5 medya:")
            for i, m in enumerate(sirali[:5], start=1):
                print(f"{i}. {m}")

        elif secim == "7":
            arsiv.kaydet()

        elif secim == "8":
            arsiv.kaydet()
            print("ðŸ‘‹ Ã‡Ä±kÄ±ÅŸ yapÄ±lÄ±yor...")
            break

        else:
            print("âš  GeÃ§ersiz seÃ§im!")


# ---------- Program BaÅŸlat ----------
if __name__ == "__main__":
    menu()
```
